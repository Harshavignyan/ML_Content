```python
from fbprophet import Prophet
import pandas as pd
```

```python
import warnings; 
warnings.simplefilter('ignore')
```

```python
df = pd.read_csv('dataset.csv')
df.head()
```

```python
df.describe()
```

```python
df.shape
```

```python
# here our data column is Time Date 
# and the y of values column is value
# in order to consider the time series model
```

```python
print(df['Product'].unique())
print(df['Store'].unique())
```

```python
df.dtypes

# we need to convert the time date column into date and time dtype
```

```python
df['Year'] = df['Time Date'].apply(lambda x: str(x)[-4:])
df['Month'] = df['Time Date'].apply(lambda x: str(x)[-6:-4])
df['Day'] = df['Time Date'].apply(lambda x: str(x)[:-6])
df['ds'] = pd.DatetimeIndex(df['Year']+'-'+df['Month']+'-'+df['Day'])
df.head()
```

```python
df.drop(['Time Date', 'Product', 'Store', 'Year', 'Month', 'Day'], axis=1, inplace=True)
df.columns = ['y', 'ds']
df.head()
```

```python
# in this case we have only one product and one store
# if there are diff products and diff stores
# use this kind of filtering

# df = df.loc[(df['Product']==2667437) & (df['Store']=='QLD_CW_ST0203')]
```

```python
# training the model
```

```python
m = Prophet(interval_width=0.95, daily_seasonality=True)
model = m.fit(df)

# interval width is the confidence interval, which is considered as 95%
# even if we dont use daily_seasonality the model is going to get built
# but to consider the seasonality factor we are taking this
# as in the dataset we are specifically talking abt every day sales
```

```python
# after this step, we perform prediction using forecasting
```

```python
future = m.make_future_dataframe(periods=100,freq='D')
forecast = m.predict(future)
forecast.head()
```

```python
plot1 = m.plot(forecast)
```

```python
plt2 = m.plot_components(forecast)
```
